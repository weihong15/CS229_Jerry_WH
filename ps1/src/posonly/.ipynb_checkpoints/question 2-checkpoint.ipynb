## 2c

$$
p(t^{(i)}=1\mid y^{(i)} = 1, x^{(i)}) = \frac{p(y^{(i)}=1\mid t^{(i)} = 1, x^{(i)})\cdot p(t^{(i)}=1\mid x^{(i)})}{p(y^{(i)}=1\mid t^{(i)} = 1, x^{(i)})\cdot p(t^{(i)}=1\mid x^{(i)})+p(y^{(i)}=1\mid t^{(i)} = 0, x^{(i)})\cdot p(t^{(i)}=0\mid x^{(i)})}
$$

Since $\forall  x, ~~ p(y^{(i)} = 1 \mid t^{(i)}=0, x^{(i)}=x) = 0,$ so the equation above becomes 1 as the numerator is the same as the denominator

$$\frac{p(y^{(i)}=1\mid t^{(i)} = 1, x^{(i)})\cdot p(t^{(i)}=1\mid x^{(i)})}{p(y^{(i)}=1\mid t^{(i)} = 1, x^{(i)})\cdot p(t^{(i)}=1\mid x^{(i)})} = 1 \implies p(t^{(i)}=1\mid y^{(i)} = 1, x^{(i)}) =1 $$


## 2d

From bayes rules we know the following is equal

$$p(t^{(i)}=1\mid y^{(i)} = 1, x^{(i)}) p(y^{(i)} = 1\mid x^{(i)}) = p(y^{(i)}=1\mid t^{(i)} = 1, x^{(i)}) p(t^{(i)} = 1\mid x^{(i)}) $$

from 2c) we know that $p(t^{(i)}=1\mid y^{(i)} = 1, x^{(i)}) =1 $ and $\forall  x, ~~ p(y^{(i)} = 1\mid t^{(i)}=1, x^{(i)}=x) = \alpha$, so the above equation becomes

$$ p(y^{(i)} = 1\mid x^{(i)}) = \alpha \cdot p(t^{(i)} = 1\mid x^{(i)}) $$

$$ \implies \begin{align}p(t^{(i)} = 1\mid x^{(i)}) = \frac{1}{\alpha}\cdot p(y^{(i)} = 1\mid x^{(i)})\label{eqn:3} \end{align}$$

## 2e

$$
\begin{align}
\mathbb{E}[h(x^{(i)})\mid y^{(i)}=1] & =  \mathbb{E}[p(y^{(i)} = 1\mid x^{(i)})\mid y^{(i)}=1]\\
& = \mathbb{E}[\alpha p(t^{(i)} = 1\mid x^{(i)})\mid y^{(i)}=1]\quad \text{by part 2d}\\
& = \alpha \quad \text{by part 2c}
\end{align}
$$
